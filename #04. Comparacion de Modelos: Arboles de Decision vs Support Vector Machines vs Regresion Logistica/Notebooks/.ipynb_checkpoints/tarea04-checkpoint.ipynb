> Copiamos y pegamos en una celda las siguientes líneas de código para cargar una tabla de datos que contiene información de personas que han solicitado tarjetas de crédito. La variable objetivo se sitúa en la columna `target` y nos cuenta si una persona ha sido apta para la tarjeta o no...

````python
import pandas as pd
            
df_base = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data',
                 na_values='?', header=None)

df_base.rename(columns={15: 'target'}, inplace=True)
df_base.head()
````

````python
pd.get_dummies(df_base)
````

````python
df = pd.get_dummies(df_base, drop_first=True)
````

````python
df.head()
````

````python
df.dropna(inplace=True)
df.sample(10)
````

````python
df.dtypes
````

````python
df.rename(columns={'target_-': 'target'}, inplace=True)
print(df)
````

# Entrenar Modelo `DecisionTreeClassifier()`

> - Posteriormente, guardar predicciones en `predicciones_dt`.

````python
from sklearn.tree import DecisionTreeClassifier
````

````python
model = DecisionTreeClassifier(max_depth=3)
````

````python
model.fit(X=explicativas, y=objetivo)
````

````python
import matplotlib.pyplot as plt
from sklearn import tree
````

````python
plt.figure(figsize=(18,12))
tree.plot_tree(decision_tree=model, feature_names=explicativas.columns, fontsize=13, filled=True);
````
